```python
import os
import pandas as pd
import zipfile
```

```python
# folder
folder = r'Data'
```

```python
import glob
fhs = glob.glob(r'E:\EC_Data\AmeriFlux\AMF_*3-5.zip')
data = {'Site':[],'Period':[],'Dataset':[]}
for fh in fhs:
    data['Site'].append(os.path.basename(fh).split('_')[1])
    data['Period'].append(os.path.basename(fh).split('_')[4])
    data['Dataset'].append('AMF')
```

```python
fhs = glob.glob(r"E:\EC_Data\OzFlux\FLX_*_beta-5.zip")
for fh in fhs:
    data['Site'].append(os.path.basename(fh).split('_')[1])
    data['Period'].append(os.path.basename(fh).split('_')[4])
    data['Dataset'].append('OzFlux')
```

```python
fhs = glob.glob(r"E:\EC_Data\ICOS\Drought2018\FLX_*_beta-3.zip")
for fh in fhs:
    data['Site'].append(os.path.basename(fh).split('_')[1])
    data['Period'].append(os.path.basename(fh).split('_')[4])
    data['Dataset'].append('ICOS-Drought')
```

```python
fhs = glob.glob(r"E:\EC_Data\ICOS\WarmWinter2020\FLX_*_beta-3.zip")
for fh in fhs:
    data['Site'].append(os.path.basename(fh).split('_')[1])
    data['Period'].append(os.path.basename(fh).split('_')[4])
    data['Dataset'].append('ICOS-WW')
```
