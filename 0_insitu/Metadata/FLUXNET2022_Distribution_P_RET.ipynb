# Calculate sites long-term annual average PCP and RET

```python
df_P = pd.read_csv('FLUXNET2022_AGERA5-PF-A_1979-2024.csv',index_col=0)
df_P
```

```python
df_E = pd.read_csv('FLUXNET2022_AGERA5-ET0-A_1979-2024.csv',index_col=0)
df_E
```

```python
def row_statistics(df):
    stats = df.apply(
        lambda row: pd.Series(
            {
                'mean': row.mean(),
                'median': row.median(),
                'std': row.std(),
                # 'sum': row.sum()
            }
        ), 
        axis=1
    )
    return stats

def calculate_statistics(df):
    stats = pd.DataFrame({
        'mean': df.mean(),
        'median': df.median(),
        'std': df.std(),
        # 'sum': df.sum()
    })
    return stats
```

```python
df_P.replace(-9999, np.nan, inplace=True)
df_PA = calculate_statistics(df_P)
df_PA.to_csv('FLUXNET2022_AGERA5-PF-A_stats.csv')
df_PA
```

```python
df_E.replace(-9999, np.nan, inplace=True)
df_EA = calculate_statistics(df_E)
df_EA.to_csv('FLUXNET2022_AGERA5-ET0-A_stats.csv')
df_EA
```

```python
from matplotlib.colors import ListedColormap, BoundaryNorm

# Function to read .clr file and extract RGB values and their corresponding values
def read_clr_file(filepath):
    colors = []
    color_values = []
    with open(filepath, 'r') as file:
        for line in file:
            if not line.startswith('#'):
                values = line.strip().split()
                color_value = int(values[0])  # Use values[0] as the color value
                r, g, b = int(values[1]), int(values[2]), int(values[3])
                colors.append((r / 255, g / 255, b / 255))  # Normalize to [0, 1] range
                color_values.append(color_value)
    return colors, color_values

# Read the .clr file
clr_filepath = 'koppen_geiger_colormap.clr'
colors, color_values = read_clr_file(clr_filepath)

# Create a custom colormap using the colors
cmap = ListedColormap(colors)

# Create a BoundaryNorm to map values to colors exactly
boundaries = [v - 0.5 for v in color_values] + [color_values[-1] + 0.5]
norm = BoundaryNorm(boundaries,  cmap.N)
```
