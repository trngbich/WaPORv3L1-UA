```python
import requests
from bs4 import BeautifulSoup

# The URL of the site to scrape
url = "https://fluxnet.org/sites/siteinfo/AU-Wac"

# Send a GET request to the site
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Function to extract information
def get_info(label):
    info = soup.find(text=label)
    return info.find_next('td').text.strip() if info else None

# Extract Vegetation IGBP and Climate Koeppen
vegetation_igbp = get_info('Vegetation IGBP:')
climate_koeppen = get_info('Climate Koeppen:')

print(f'Vegetation IGBP: {vegetation_igbp}')
print(f'Climate Koeppen: {climate_koeppen}')
```
