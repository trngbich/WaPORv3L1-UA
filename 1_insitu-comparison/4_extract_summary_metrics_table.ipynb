# Calculate and extract metrics table




```python
products = ['WaPOR_100m','PML','MODIS','SSEBop']
metrics = ['r', 'Bias', 'RMSE', 'PBias','N']
```

```python
df_
```

```python
for i,product in enumerate(products):
    csv = f'Data/222sites_metrics_{product}.csv'
    df_ = pd.read_csv(csv,index_col=0)
    df_ = df_[['Site','r','rmse','bias','pbias','N','EC average']]
    df_ = df_.rename(columns={
        'EC average': f'ECavg_{product}', 
        'Site' : 'Station',
        'N': f'N_{product}', 
        'r': f'r_{product}', 
        'rmse': f'RMSE_{product}', 
        'bias': f'Bias_{product}', 
        'pbias': f'PBias_{product}'})
    df_ = df_.set_index('Station')
    if i == 0:
        df = df_
    else:
        df = df.join(df_, how='left')
```
