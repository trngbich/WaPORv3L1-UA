```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Load the single Excel file
file_path = r"Data\222sites_metrics_ET_products.xlsx"
df = pd.read_excel(file_path)
df
```

```python
# Define dataset and metric names
datasets_order = ['WaPOR_100m', 'SSEBop', 'MODIS','PML']
datasets_labels = ['WaPORv3L1', 'SSEBopv61', 'MOD16A2GF','PMLv2']
metrics = ['r', 'Bias', 'RMSE', 'PBias']

# Best values for reference lines
best_values = {
    'r': 1.0,
    'Bias': 0.0,
    'PBias': 0.0,
    'RMSE': 0.0
}

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(10, 4))
labels = ['a', 'b', 'c', 'd']

# Loop through each metric
for ax, metric, label in zip(axs.flatten(), metrics, labels):
    # Extract relevant columns for the current metric
    metric_cols = [f"{metric}_{ds}" for ds in datasets_order]
    metric_data = df[metric_cols].copy()
    metric_data.columns = datasets_order  # Rename columns to dataset names

    # Clip values for clean visualization
    if metric == 'PBias':
        metric_data = metric_data.clip(lower=-100, upper=100)
    elif metric == 'Bias':
        metric_data = metric_data.clip(lower=-2, upper=3)
    elif metric == 'RMSE':
        metric_data = metric_data.clip(lower=0, upper=3)

    # Plot boxplot
    sns.boxplot(data=metric_data, ax=ax, palette="Set2", linewidth=0.9, width=0.5)

    # Calculate mean values
    means = metric_data.mean()
    #calculate median values
    medians = metric_data.median()
    # Print the medians in console
    print(f"\nMetric: {metric}")
    for ds, val in zip(datasets_order, medians):
        print(f"  {ds}: {val:.2f}")

    # Plot mean as dashed line inside each box (width=0.5 â†’ half=0.25)
    for i, mean_val in enumerate(means):
        ax.hlines(mean_val, i - 0.25, i + 0.25, colors='grey', linestyles='--', linewidth=1.2, zorder=3)

    # Title formatting
    if metric == 'r':
        ax.set_ylabel(r'$\rho$ [-]', fontsize=12)
    elif metric == 'Bias':
        ax.set_ylabel('Bias [mm/day]', fontsize=12)
    elif metric == 'RMSE':
        ax.set_ylabel('RMSE [mm/day]', fontsize=12)
        ax.set_yticks([0, 1, 2, 3])
    elif metric == 'PBias':
        ax.set_ylabel('PBias [%]', fontsize=12)

    # Formatting
    # ax.set_ylabel('')
    ax.set_xticklabels(datasets_labels, rotation=90, fontsize=10)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.axhline(best_values[metric], color='grey', linestyle='--', linewidth=0.8, label='Optimal value')
    ax.grid(False)
    ax.text(0.5, 1.05, f'({label})', transform=ax.transAxes, fontsize=12, va='top', ha='center')

# Final layout adjustments
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

# Save figure (optional)
plt.savefig(r"Figure/5-7.png", bbox_inches='tight', dpi=300)
plt.show()
```
